```python
df=pd.read_csv('cleaned_titanic.csv')
df.head()
```

```python
# Mean of numeric columns
print("Mean:\n", df.mean(numeric_only=True))

# Median of numeric columns
print("\nMedian:\n", df.median(numeric_only=True))

# Standard Deviation of numeric columns
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

```python
df[['Age', 'Fare']].hist(figsize=(10, 4), bins=10, color='skyblue', edgecolor='black')
plt.suptitle("Histogram of Age and Fare")
plt.show()


```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot of Age and Fare")
plt.show()
```

# 🔍 **Patterns and Trends**

- Most passengers were aged 20–40.
- Females had higher survival rate.
- Fare has many outliers (seen in boxplot).


```python

plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived')
plt.suptitle("Pairplot of Age, Fare, and Survived", y=1.02)
plt.show()
```

```python

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```


** Feature-Level Inferences from Visuals**
### 🔹 Sex
- From the bar plot, we see that **females had a higher survival rate** than males.
- About 74% of females survived, while only ~19% of males survived.

### 🔹 Pclass (Passenger Class)
- **1st class passengers had better survival chances** than 2nd and 3rd class.
- 3rd class passengers had the **lowest survival rate**.

### 🔹 Fare
- **Higher fare-paying passengers survived more** (seen in boxplot and pairplot).
- Most passengers paid less than 50.
- Some passengers paid extremely high fare (outliers).

### 🔹 Age
- Most passengers were between **20–40 years** (seen in histogram).
- Few very old passengers (above 70) are anomalies.
- **Children (Age < 10)** had slightly better survival.

### 🔹 Embarked
- Majority of passengers boarded from port **'S'**.
- Passengers from port **'C'** had slightly better survival rate.

### 🔹 SibSp / Parch (Family onboard)
- Passengers with **1 or 2 family members** had better survival.
- Solo travelers and large families had **lower survival chances**.
